# IPL Dataset Mini Project

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
ipl=pd.read_csv('IPL_Dataset.csv')
ipl
```

### Q1. What is the maximum number of matches played by an individual player in a season?

```python
max_matches = ipl.groupby('Player')['Mat'].max()
max_player = max_matches.idxmax()
max_matches_count = max_matches.max()
print(f"{max_player} played max {max_matches_count} matches in a season.")
```

### Q2. What is the highest average of a player in a season?

```python
highest_average = ipl.groupby('Player')['Avg'].max()
max_avg_player = highest_average.idxmax()
max_avg = highest_average.max()
print(f" Highest average in a season is {max_avg} by {max_avg_player}.")
```

### Q3. What is the average number of matches played by the players in a season?

```python
average_matches_played = ipl['Mat'].mean()
average_matches_played
print(f"The average number of matches played by the players in a season is {average_matches_played}")
```

### Q4. Please check the correlation between all the features.

```python
numeric_data = ipl.select_dtypes(include=['float64','int64'])
correlation = numeric_data.corr()
print("Correlation between features: ")
print(correlation)
```


### Q5. Check the list of players who has an average greater than 50 as well strike rate above 120.

```python
average = ipl[(ipl['Avg']> 50) & (ipl['SR'] > 120)]
index_order = average[['Player','Avg','SR']].reset_index(drop=True)
print('Players who have an average greater than 50 and strike rate below 120 :')
print(index_order)
```

### Q.6: Please check the list of players who has an average greater than 40 and balls faced above 100?

```python
average = ipl[(ipl['Avg']> 40) & (ipl['BF'] > 100)]
index_order = average[['Player','Avg','BF']].reset_index(drop=True)
print('Players who have an average greater than 40 and faced balls below 100 :')
print(index_order)
```

### Q.7: Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130?

```python
list_of_player = ipl[(ipl['Runs'] > 300) & (ipl['Avg'] > 45) & (ipl['SR'] > 130)]
index_order =  list_of_player[['Player', 'Avg', 'BF']].reset_index(drop=True)
print('The list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130:')
print(index_order)
```

### Q-8. Players who scroed alteast one century in this season.

```python
century = ipl[(ipl['100']>=1)]
index_order =century[['Player','100']].reset_index(drop=True)
print('Players who scroed alteast one century in this season: ')
print(index_order)
```

### Q.9: Players who scored atleast 4 half centuries in this season?

```python
fifty = ipl[ipl['50'] >= 4]
index_order = fifty[['Player', '50']].sort_values(by='50', ascending=False).reset_index(drop=True)
print('The list of players who scored at least four half century in this season:')
print(index_order)
```

### Q.10: Check the list of players who hit more than 45 boundaries and more than 10 sixes in this season?¶

```python
boundaries = ipl[(ipl['4s'] > 45) & (ipl['6s'] > 10)]
index_order = boundaries[['Player', '4s', '6s']].sort_values(by=['4s', '6s'], ascending=False).reset_index(drop=True)
print('The list of players who hits more than 45 boundaries and more than 10 sixes in this season:')
print(index_order)
```

### Q.11: Plot a histogram of number of matches played in a season by players?

```python
plt.hist(ipl['Mat'], bins=10, edgecolor='black', color='blue')
plt.title('Number of Matches Played in a Season by Players')
plt.xlabel('Number of Matches')
plt.ylabel('Player')
plt.show()
```

### Q-12. Plot the histogram of balls faced by players.

```python
plt.hist(ipl['BF'], bins=8, edgecolor='black', color='lightgreen')
plt.title('Balls Faced by Players')
plt.xlabel('Number of Matches')
plt.ylabel('Player')
plt.show()
```

### Q.13: Top 20 players with most runs in a season?

```python
most_runs = ipl.sort_values(by='Runs', ascending=False).head(20)
print('Top 20 players with most runs in a season: ')
print(most_runs[['Player', 'Runs']])
```

### Q.14: Bar chart for top 20 players with best strike rate in a season?

```python
top_sr_players= ipl.nlargest(20,'SR')
plt.figure(figsize=(10,7))
plt.bar(top_sr_players['Player'],top_sr_players['SR'], edgecolor="black", color='orange')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.title('Top 20 Players with Best Strike Rate in a Season')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

### Q.15: Top 30 players with highest number of sixes?

```python
highest_sixes = ipl.nlargest(30, '6s')
highest_sixes_reset = highest_sixes.reset_index(drop=True)
print('Top 30 players with Highest number of Sixes:')
print(highest_sixes_reset[['Player', '6s']])
```

### Q.16: Scatter plot of runs scored by a player v/s balls faced in a season.?

```python
plt.scatter(ipl['Runs'], ipl['BF'], alpha=0.5, edgecolor='black', color='yellow')
plt.title('Runs scored by a player vs balls faced in a season')
plt.xlabel('Runs Scored')
plt.ylabel('Balls Faced')
plt.show()
```

### Q.17: KDE plot for averages on a season?

```python
sns.kdeplot(ipl['Avg'], fill=True, edgecolor='black', color='red')
plt.title('KDE plot for averages on a season')
plt.xlabel('Average')
plt.ylabel('Density')
plt.show()
```
